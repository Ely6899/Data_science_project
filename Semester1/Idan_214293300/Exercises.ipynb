```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = int(input("Enter any number: "))

to_string = {10:"A", 11:"B", 12:"C", 13:"D", 14:"E", 15:"F"}
for i in range(10):
    to_string[i] = str(i)

copy = x
for i in [2, 8, 16]:
    new_x = ""
    x = copy
    while(x != 0):
        remainder = x%i
        new_x += to_string[remainder]
        x = x//i
    print("Base " + str(i) + ": " + new_x[::-1])
```
