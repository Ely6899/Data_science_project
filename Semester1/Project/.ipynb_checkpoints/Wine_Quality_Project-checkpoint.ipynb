```python
fig, ax = plt.subplots(3, 1, figsize=(15,10), sharex = True)
fig.suptitle("Acids (Not scaled)", fontsize = 30)


#create subplots
ax[0].plot(data['fixed acidity'], color = 'red')
ax[0].set_ylabel('fixed acidity', fontsize = 15)

ax[1].plot(data['volatile acidity'], color = 'green')
ax[1].set_ylabel('volatile acidity', fontsize = 15)

ax[2].plot(data['citric acid'], color = 'blue')
ax[2].set_ylabel('citric acid', fontsize = 15)


ax[0].title.set_text("Fixed acidity")
ax[1].title.set_text("Volatile acidity")
ax[2].title.set_text("Citric acid")

fig.text(0.55, -0.01, 'sample ID', ha='center', fontsize = 20)
fig.tight_layout(h_pad = 3)
```

```python
data.corr()
```

```python
X = data.iloc[:, :11]
y = data.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.90, random_state = 0)
```

```python
X_train
```

```python
y_train
```

```python
# training a linear SVM classifier
from sklearn.svm import SVC
svm_model_rfb = SVC(kernel = 'rbf', C = 1).fit(X_train, y_train)
svm_predictions = svm_model_rfb.predict(X_test)
```

```python
# model accuracy for X_test 
accuracy_rfb = svm_model_rfb.score(X_test, y_test)
accuracy_rfb
```

```python
# creating a confusion matrix
from sklearn.metrics import confusion_matrix
cm_rfb = confusion_matrix(y_test, svm_predictions)
cm_rfb
```

```python
svm_model_linear = SVC(kernel = 'linear', C = 1).fit(X_train, y_train)
svm_predictions = svm_model_linear.predict(X_test)
```

```python
accuracy_linear = svm_model_linear.score(X_test, y_test)
accuracy_linear
```

```python
cm_linear = confusion_matrix(y_test, svm_predictions)
cm_linear
```
